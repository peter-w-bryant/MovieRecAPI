```python
import pandas as pd

df_movies = pd.read_csv('C:/Users/pwbry/OneDrive/Desktop/Documents/Projects/MovieRecAPI/ml-latest-small/movies.csv')
df_ratings = pd.read_csv('C:/Users/pwbry/OneDrive/Desktop/Documents/Projects/MovieRecAPI/ml-latest-small/ratings.csv')
df_merged = pd.merge(df_movies, df_ratings, on='movieId', how='inner')
df_merged.head()
```

 Create a dataframe which maps individual user ratings into rows against each movie as a column and drop movies that had fewer than 8 ratings.

```python
df = df_merged.pivot_table(index='userId', columns='title', values='rating')
# Keep only movies that had at least 8 ratings
df = df.dropna(thresh=8, axis=1)
df.fillna(0, inplace=True)
df.head()
```

Uses the Pearson correlation to calculate the similarity between movies!

The Pearson method treats each movie (i.e. column) as a vector containing user rating values and determines how close/similar a movie is to the other ones.

```python
df_similarity = df.corr(method='pearson')
#Store the data for later to be used in building the API
df_similarity.to_csv('movie_similarity.csv')
df_similarity.head()
```

List of 10 movies that are the most similar to 'Harry Potter and the Goblet of Fire (2005)'!

```python
movieLiked = 'Harry Potter and the Goblet of Fire (2005)'
similarityScores = df_similarity[movieLiked]
similarityScores.sort_values(ascending=False)[1:10] # Display 10 most similar movies
```
